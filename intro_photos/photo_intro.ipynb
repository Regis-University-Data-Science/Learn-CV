### Loading Images with OpenCV  
1. OpenCV’s `imread()` function loads an image as a NumPy array:  
   ```python
   image = cv2.imread("path_to_image.jpg")
   ```

2. The shape of the image provides the following information:  
   * **Height**: Number of rows (pixels) from top to bottom  
   * **Width**: Number of columns (pixels) from left to right  
   * **Channels**: Number of color layers (typically 3 for color images)  
   * OpenCV loads images in **BGR** (Blue, Green, Red) format by default. 
> This differs from most image libraries such as Matplotlib or PIL, which use the more common **RGB** (Red, Green, Blue) format. To display the image correctly using these libraries, it’s standard practice to convert BGR to RGB using:  
     ```python
     image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
     ```


````python
print(os.listdir())
image = cv2.imread('dog.jpg')
print(image.shape)
# Check if image was loaded successfully
if image is None:
    print("Error: Image not found or failed to load.")
else:
    # Get shape
    height, width, channels = image.shape
    print(f"Image shape: {image.shape}")
    print(f"Height: {height} pixels")
    print(f"Width: {width} pixels")
    print(f"Channels: {channels}")
    
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
````

### Displaying an Image with Matplotlib  
1. After converting the image to RGB format, we can display it using Matplotlib:  
   ```python
   image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
   plt.imshow(image_rgb)
   plt.title("Shanti Bday Party")
   plt.show()
   ```
2. The `imshow()` function renders the image, and `title()` adds a caption above it.  
3. axis('off')` hides the x and y axes.  
> This step is useful for visual inspection and verifying that the image is loaded and displayed correctly.


````python
plt.imshow(image_rgb)
plt.title("Shanti")
plt.show()
````

````python
plt.imshow(image_rgb)
plt.title("Shanti")
plt.axis('off')
plt.show()
````

## Exporting an Image with OpenCV
Exporting an image means saving it from your program to a file on disk so it can be stored, shared, or opened later. <br>
cv2.imread() → Reads an image file into a NumPy array so it can be processed. <br>
cv2.imwrite() → Saves the processed (or original) image array to a new file. <br>
The output file format is determined by the file extension you provide.    <br>



````python
img = cv2.imread("dog.jpg")
cv2.imwrite("shanti_export.jpg", img)
image = cv2.imread("shanti_export.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Shanti Export")
plt.show()
````
