### Reshaping an Image  
1. Reshaping means changing the structure of the image array using NumPy’s `reshape()` method:  
   ```python
   reshaped = image.reshape(new_height, new_width, channels)
   ```
2. However, in computer vision, this is **rarely used directly on images**, because:  
   * Images have a fixed pixel layout — reshaping without changing content leads to distortion.  
   * `reshape()` only rearranges pixel values in memory without preserving spatial structure.  
   * It’s typically used for flattening images (e.g., `image.reshape(-1)`), or feeding into ML models, not for visual display.  

> To change image size visually (for display or processing), we use **resizing**, not reshaping:  
   ```python
   resized = cv2.resize(image, (new_width, new_height))
   ```
> Resizing preserves spatial layout and interpolates pixel values, making it appropriate for zoom, scaling, or preparing inputs for models.
