```python
print(os.listdir())
video_path = "pikes_peak.mp4"
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
else:
    # Read the first frame
    ret, frame = cap.read()
    if ret:
        # Convert from BGR to RGB (OpenCV uses BGR by default)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame_rgb)
        plt.title("First Frame")
        plt.axis("off")
        plt.show()
    else:
        print("Error: Couldn't read a frame from the video.")
        
cap.release()
    
```

```python
frame_rotated = cv2.rotate(frame, cv2.ROTATE_180)

# Convert to RGB for displaying with matplotlib
frame_rgb = cv2.cvtColor(frame_rotated, cv2.COLOR_BGR2RGB)

# Show the rotated frame
plt.imshow(frame_rgb)
plt.title("First Frame (Corrected)")
plt.axis("off")
plt.show()
```
